```python
import re

programs = ["lu"]

for p in programs:
    for i in [1, 2, 3]:
        # Define regex patterns to extract required information
        active_processes_pattern = r'Number of active processes:\s+(\d+)'
        #threads_per_process = r'Total number of threads:\s+\d+\s+\(\s+(\d+)\.\d+ threads/process\)'
        threads_pattern = r'Total number of threads:\s+(\d+)'
        speedup_pattern = r'Calculated speedup =\s+([\d.]+)'
        time_pattern = r'Time in seconds =\s+([\d.]+)'

        # Open file for reading
        with open(f'ludata/lu{i}.txt', 'r') as f:

            # Initialize variables to store extracted information
            active_processes = None
            threads_pp = None
            threads = None
            speedup = None
            time = None

            # Iterate over each line in the file
            for line in f.readlines():

                # Extract information from line using regex
                match = re.search(active_processes_pattern, line)
                if match:
                    active_processes = match.group(1)
                    continue
                
                '''
                match = re.search(threads_per_process, line)
                if match:
                    threads_pp = match.group(1)
                    continue
                '''
                match = re.search(threads_pattern, line)
                if match:
                    threads = match.group(1)
                    continue

                match = re.search(speedup_pattern, line)
                if match:
                    speedup = match.group(1)
                    continue

                match = re.search(time_pattern, line)
                if match:
                    time = match.group(1)
                    continue
                

                # Create table if all required information is available
                if active_processes  and threads and speedup and time:
                    with open(f'ludata/out{i}.txt', 'a') as f:
                        f.write(f"{time:>14} \n")

                    # Reset variables to None to start processing next set of information
                    active_processes = None
                    #threads_pp = None
                    threads = None
                    speedup = None
                    time = None
```
